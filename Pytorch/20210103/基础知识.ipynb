## 创建Tensor

```python
# 第一种方式 从numpy引入
a = np.array([2,3.3])
a = torch.from_numpy(a)

b = np.ones([2,3])
b = torch.from_numpy(b)

print(a)
print(b)
```

```python
# 第二种 是从list导入
a = torch.tensor([2.,3.2])

b = torch.FloatTensor([2.,3.2])

c = torch.tensor([[2.,3.2],[1.,3.]])

print(a)
print(b)
print(c)
```

### 随机初始化  rand/rand_like,randint
* [0,1]
* [min,max]
* _like

```python
a = torch.rand(3,3)
print(a)
a = torch.rand_like(a)
print(a)
a = torch.randint(1,10,[4,4])
print(a)
```
